**Department** **(work with columns)**

```python
# Plotting a bar chart for Department and its count
bx = sns.set(rc = {'figure.figsize': (15, 5)})
bx = sns.countplot(x = 'Department', data = df)

for bars in bx.containers:
    bx.bar_label(bars)
```

```python
# Plotting a bar chart for department vs Employees
dep_emp = df.groupby(['Department'], as_index= False)['EmployeeNumber'].sum().sort_values(by = 'EmployeeNumber', ascending = False)

bx = sns.barplot(x = 'Department', y = 'EmployeeNumber', data = dep_emp)

for bars in bx.containers:
    bx.bar_label(bars)
```

**AgeGroup**

```python
# Plot a bar chart of age group vs gender
bx = sns.countplot(data = df, x = 'AgeGroup', hue = 'Gender')

for bars in bx.containers:
    bx.bar_label(bars)
```

```python
# Plot age group vs employeecount
age_emp = df.groupby(['AgeGroup'], as_index = False)['EmployeeCount'].sum().sort_values(by = 'EmployeeCount', ascending = False)

bx = sns.barplot(x = 'AgeGroup', y = 'EmployeeCount', data = age_emp)

for bars in bx.containers:
    bx.bar_label(bars)


```

**JobRole**

```python
# plotting a graph between jobs and employeecount
Job_emp = df.groupby(['JobRole'], as_index= False)['EmployeeCount'].sum().sort_values(by = 'EmployeeCount', ascending = False).head(7)

bx = sns.set(rc= {'figure.figsize': (20, 5)})
bx = sns.barplot(x= 'JobRole', y= 'EmployeeCount', data = Job_emp)

for bars in bx.containers:
    bx.bar_label(bars)
```
